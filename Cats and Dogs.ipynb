```python
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

DATADIR = "/home/tony/Documents/deepLearning/Cats and Dogs/kagglecatsanddogs_3367a/PetImages/"
CATEGORY = ["Dog","Cat"]

for category in CATEGORY:
    path = os.path.join(DATADIR,category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img), 0)
        plt.imshow(img_array,cmap = 'gray')
        plt.show()
        break
    break
```

```python
print(img_array.shape)
```

```python
IMG_SIZE = 50
newArray = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(newArray, cmap ='gray')
plt.show()
```

```python
training_list =[]
def create_training_list():
    for category in CATEGORY:
        path = os.path.join(DATADIR,category)
        class_num = CATEGORY.index(category)
        for img in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path,img),0)
                newArray = cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_list.append([newArray,class_num])
            except Exception as e:
                pass

create_training_list()
```

```python
print(len(training_list))
```

```python
import random
random.shuffle(training_list)
```
